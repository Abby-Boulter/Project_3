```python

def Merge(dict1, dict2, dict3, dict4, dict5, dict6, dict7, dict8, dict9, dict10, dict11):
    # create a new dictionary by merging the items of the two dictionaries using the union operator (|)
    merged = dict(dict1.items() | dict2.items() | dict3.items() | dict4.items() | dict5.items() | dict6.items() | dict7.items() | dict8.items() | dict9.items() | dict10.items() | dict11.items())
    # return the merged dictionary
    return merged

Apache_XHD = Merge(Apache_XHD2011, Apache_XHD2012, Apache_XHD2013, Apache_XHD2014, Apache_XHD2015, Apache_XHD2016, Apache_XHD2017, Apache_XHD2018, Apache_XHD2019, Apache_XHD2020, Apache_XHD2021)
Apache_XHD
```
