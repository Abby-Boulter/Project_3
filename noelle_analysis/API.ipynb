# 1. Retrieve data from the EPHTN

```python
# Dependencies
import requests
import json

# URL for GET requests to retrieve data
version = "v1"
geographicTypeId = 2 # county level, 1 is state
contentAreaIdFilter = 35
returnType = "json"
# indicatorIdFilter = 19

# pull content areas
content_pull = f"https://ephtracking.cdc.gov/apigateway/api/{version}/contentareas/{returnType}"
```

```python
content_pull
```

```python
# Run a request to endpoint and convert result to json
geo_data = requests.get(content_pull).json()

# Print the json (pretty printed)
print(json.dumps(geo_data, indent=4, sort_keys=False))
```
