```python
# Import dependencies
import pandas as pd
import numpy as np
pd.set_option('max_colwidth', 400)
```

### Extract the project 3 Data

### ED Visit Data

```python
# Read the ED Visits data into a Pandas DataFrame
ed_visits_df = pd.read_csv('../data/AZ-EDVisits.csv')
ed_visits_df.head()
```

```python
# Get a brief summary of the ED visits DataFrame.
ed_visits_df.info()
```

```python
# Get the ed_visits_df column names.
columns = list(ed_visits_df.columns)
print(columns)
```

```python
#Get the unique value counts in the columns
for i in columns: 
    print(i, ed_visits_df[i].nunique())
```

```python
#Pull Necessary columns and rename
ed_visits_df = ed_visits_df[['Name', 'Year', 'GeogID', 'Value']]
ed_visits_df.rename(columns={'Name': 'county', 'Year': 'year', 'GeogID': 'county_id', 'Value': 'number_of_visits'}, inplace=True)
ed_visits_df.head()
```

```python
# Export the cleaned DataFrame as a CSV file. 
ed_visits_df.to_csv("../resources/cleaned-AZ-EDVisits.csv", encoding='utf8', index=False)
```
