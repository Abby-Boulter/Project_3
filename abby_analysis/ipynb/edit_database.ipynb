```python
# Import SQL Alchemy
from sqlalchemy import create_engine

# Import and establish Base for which classes will be constructed 
from sqlalchemy.ext.declarative import declarative_base
Base = declarative_base()

# Import modules to declare columns and column data types
from sqlalchemy import Column, Integer, String, Float, Boolean

from pathlib import Path
import pandas as pd
```

```python
path = Path('hri.sqlite')
```

```python
engine = create_engine(f'sqlite:///hri.sqlite')
```

```python
tables = (engine.execute('''SELECT name FROM sqlite_master WHERE type="table"'''))
print(tables.fetchall())
```

```python
cons = pd.DataFrame(engine.execute('''SELECT * FROM consolidate'''))
cons.head()
```

```python
cons = cons.reset_index()
cons.head()
```

```python
cons = cons.rename(columns={'index': 'id'})
cons.head()
```

```python
# Export the cleaned DataFrame as a CSV file. 
cons.to_csv("hri.csv", encoding='utf8', index=False)
```
